```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# Readign data
df=pd.read_csv("titanic.csv")
```

```python
#Create dummy variables ( for categorical variables)
cc=pd.get_dummies(df["Embarked"])
```

```python
cc
```

```python
X=pd.concat([df,cc],axis=1)
```

```python
X
```

```python
X=X.drop(['Embarked'],axis=1)
```

```python
X
```

```python
#Create dummy variables ( for categorical variables)
gg=pd.get_dummies(X["Sex"])
X=pd.concat([X,gg],axis=1)
X=X.drop(['Sex'],axis=1)
```

```python
#Create dummy variables ( for categorical variables)
kk=pd.get_dummies(X["Pclass"], prefix='origin')
X=pd.concat([X,kk],axis=1)
X=X.drop(['Pclass'],axis=1)
```
